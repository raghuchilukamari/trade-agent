# 1. Rocket stocks

Rocket stocks follow below trend (only works in bullish market) 
1. Hits all time highs everyday
    * At ATHs all buyers are happy
    * short sellers also will be forced to buy when ATHs hit consistently 
    * panicked buying kicks in 
2. Surge in Daily volume and stock gaps occur
    * check stocks that are trading at 4 - 40x daily averge volume
3. Stock trades above 50 day MA
4. 50 day MA > 200 day MA
5. Market cap < 4B 
6. Float is 20% or less of Total shares. 
    * Float is total available shares to trade
7. Short interest > 10% 
8. Check for bullish market - SPY trading above 50 MA and 50MA > 200 MA

```python
def timestamp_to_date(ts):
    try:
        date = datetime.fromtimestamp(ts).strftime('%Y-%m-%d')
    except Exception as e:
        date = datetime.strptime("9999-9-9", "%Y-%m-%d").date()
    return date
        
stock_info_df['previousMonthShortDate'] = stock_info_df['sharesShortPreviousMonthDate'].apply(lambda x: timestamp_to_date(x))
stock_info_df['floatPercent'] = (stock_info_df['floatShares']/stock_info_df['sharesOutstanding'])*100
stock_info_df['marketCap_in_B'] = stock_info_df['marketCap'] / 1000000000
stock_info_df['marketCap_lt_4B'] = np.where((stock_info_df['marketCap_in_B'].notna()) & (stock_info_df['marketCap_in_B'] <= 4), 1, 0)
stock_info_df['floatPercent_lt_20pct'] = np.where((stock_info_df['floatPercent'].notna()) & (stock_info_df['floatPercent'] <= 20),1,0)


stockInfo = stock_info_df[['sector','industry','beta','dateShortInterest','previousMonthShortDate','floatPercent','shortPercentOfFloat','marketCap_lt_4B','floatPercent_lt_20pct',]]
stockInfo['shortInterest_gt_10'] = np.where((stock_info_df['shortPercentOfFloat'].notna()) & (stockInfo['shortPercentOfFloat'] >= 10),1,0)
stockInfo['shortInterestDate'] = stockInfo['dateShortInterest'].apply(lambda x: timestamp_to_date(x))

display(stockInfo)
```

```python
import requests, nltk
from nltk.sentiment import SentimentIntensityAnalyzer


def sentiment_analysis(symbol):

    # Fetch StockTwits messages for the stock symbol
    url = f"https://api.stocktwits.com/api/2/streams/symbol/{symbol}.json?limit={limit}"
    response = requests.get(url)
    data = response.json()

    # Extract messages from the response
    messages = [msg['body'] for msg in data['messages']]

    # Initialize the Sentiment Intensity Analyzer
    sia = SentimentIntensityAnalyzer()

    # Perform sentiment analysis on each message
    sentiments = []
    for message in messages:
        sentiment_score = sia.polarity_scores(message)
        sentiment = 'Positive' if sentiment_score['compound'] >= 0 else 'Negative'
        sentiments.append(sentiment)

    # Count the number of positive and negative sentiments
    positive_count = sentiments.count('Positive')
    negative_count = sentiments.count('Negative')
    
    return (positive_count/len(messages))*100
```

```python
#from pandas_datareader import data
from ta.volatility import BollingerBands
from ta.momentum import RSIIndicator, ROCIndicator
from ta.trend import SMAIndicator, MACD
from time import sleep

stock_data_df = pd.DataFrame()
stock_rsi_df = pd.DataFrame()

for stock in tickers:
    end_date = datetime.now()
    sleep(1)
    start_date = end_date - timedelta(days=5*365)
    stock_data = yf.download(tickers=stock, start=start_date, end=end_date, interval='1d')
    stock_data['Ticker'] = stock
        
    # shift by 1 to get previous close and volumes
    stock_data['prev_Close'] = stock_data.loc[:, 'Close'].shift(1)
    stock_data['prev_Volume'] = stock_data.loc[:, 'Volume'].shift(1)

    # 50,200 SMAs
    stock_data['50SMA'] = stock_data['Close'].rolling(50).mean()
    stock_data['200SMA'] = stock_data['Close'].rolling(200).mean()

    # 50 day avg volume
    stock_data['50SMA_vol'] = stock_data['prev_Volume'].rolling(50).mean()


    # Bollinger bands
    indicator_bb = BollingerBands(close=stock_data["Close"], window=20, window_dev=2)
    stock_data["bb_upperband"] = indicator_bb.bollinger_hband()
    stock_data["bb_lowerband"] = indicator_bb.bollinger_lband()
    stock_data['bb_lb_cross'] = indicator_bb.bollinger_lband_indicator()
    stock_data['bb_ub_cross'] = indicator_bb.bollinger_hband_indicator()

    print(stock_data)
    
    # MACD
    macd = MACD(close=stock_data["Close"],window_slow=26,window_fast=12,window_sign=9)
    stock_data['macd_line'] = macd.macd()
    stock_data['macd_hist'] = macd.macd_diff()
    stock_data['macd_signal'] = macd.macd_signal()
    
    stock_data['macd_trend'] = np.where(stock_data['macd_hist'] > 0, 'Uptrend', 'Downtrend')
    stock_data['macd_trend_reversal'] = stock_data['macd_trend'].ne(stock_data['macd_trend'].shift())
    
    stock_data['uptrend_days'] = 0
    stock_data['downtrend_days'] = 0

    uptrend_count = 0
    downtrend_count = 0

    for i, row in stock_data.iterrows():
        if row['macd_trend_reversal']:
            uptrend_count = 0
            downtrend_count = 0

        if row['macd_trend'] == 'Uptrend':
            uptrend_count += 1
            stock_data.at[i, 'uptrend_days'] = uptrend_count
        elif row['macd_trend'] == 'Downtrend':
            downtrend_count += 1
            stock_data.at[i, 'downtrend_days'] = downtrend_count
    
    #ROC
#     stock_data['roc_monthly'] = ROCIndicator(stock_data['Close'], window=21).roc()
#     stock_data['roc_qrtr'] = ROCIndicator(stock_data['Close'], window=63).roc()
#     stock_data['roc_semi'] = ROCIndicator(stock_data['Close'], window=125).roc()
#     stock_data['roc_yr'] = ROCIndicator(stock_data['Close'], window=252).roc()

#     stock_data['roc'] = (ROCIndicator(stock_data['Close'], window=21).roc() + ROCIndicator(stock_data['Close'], window=63).roc() +
#                         ROCIndicator(stock_data['Close'], window=125).roc() + ROCIndicator(stock_data['Close'], window=252).roc())
    
#     roc_percentiles = stock_data['roc'].quantile([0.01,0.05,0.1, 0.9,0.95,0.99]).T
#     roc_oversold = roc_percentiles.loc[0.01]
#     roc_overbought = roc_percentiles.loc[0.99]

#     # print(roc_percentiles)
#     stock_data['roc_buy_signal'] = ((stock_data['roc'] < roc_oversold) & (stock_data['roc'].shift(1) >= roc_oversold))
#     stock_data['roc_sell_signal'] = ((stock_data['roc'] > roc_overbought) & (stock_data['roc'].shift(1) <= roc_overbought))

    
#     #RSI
#     rsi = RSIIndicator(close=stock_data["Close"])
#     stock_data['rsi'] = rsi.rsi()
#     rsi_percentiles = stock_data['rsi'].quantile([0.01,0.05,0.1, 0.9,0.95,0.99]).T
#     stock_rsi_df[stock] = rsi_percentiles
    
#     # print(rsi_percentiles)
#     # Get dynamic oversold and overbought levels
#     rsi_oversold = rsi_percentiles.loc[0.05]
#     rsi_overbought = rsi_percentiles.loc[0.95]
    
#     stock_data['rsi_buy_signal'] = ((stock_data['rsi'] < rsi_oversold) & (stock_data['rsi'].shift(1) >= rsi_oversold))
#     stock_data['rsi_sell_signal'] = ((stock_data['rsi'] > rsi_overbought) & (stock_data['rsi'].shift(1) <= rsi_overbought))
    

    # Volume surge indicator
    stock_data['volChange'] = (stock_data['Volume']/stock_data['50SMA_vol']).round(2)
    stock_data['volSurgeInd'] = np.where(stock_data['Volume']/stock_data['50SMA_vol'] > 4, 1, 0)
    
    # 52 week high & low
    stock_data['is52wkHigh'] = np.where(stock_data['Close'] >= stock_data['Close'].rolling(window=252).max(),1, 0)
    stock_data['is52wkLow'] = np.where(stock_data['Close'] <= stock_data['Close'].rolling(window=252).min(),1, 0)
    stock_data['is30DayHigh'] = np.where(stock_data['Close'] >= stock_data['Close'].rolling(window=30).max(),1, 0)
    stock_data['is21DayHigh'] = np.where(stock_data['Close'] >= stock_data['Close'].rolling(window=21).max(),1, 0)
    stock_data['is9DayHigh'] = np.where(stock_data['Close'] >= stock_data['Close'].rolling(window=9).max(),1, 0)
    
    stock_data['50gt200'] = np.where(stock_data['50SMA'] >= stock_data['200SMA'], 1, 0)
    # stock_data['50gt200Prev'] = stock_data.loc[:,'50gt200'].astype('int').shift(1)
    
    stock_data['tradingabv50SMA'] = np.where(stock_data['Close'] > stock_data['50SMA'],1,0)
    # stock_data['tradingabv50SMAtrend'] = stock_data['tradingabv50SMA'].rolling(window=252).sum()
    # stock_data['50gt200trend'] = stock_data['50gt200'].rolling(window=252).sum()
    
    # stock gap
    stock_data['gap_pct'] = round(((stock_data['Open'] - stock_data['prev_Close'])/stock_data['prev_Close'])*100,2)
    
    # 30 day high
    

    
    stock_data = stock_data.dropna()
    stock_data = stock_data.round(2)
    
    stock_data_df = stock_data_df.append(stock_data)
    
    
# display(stock_data_df.sort_values(['Ticker','Date'], ascending=False))
stock_data_df.sort_values(['Ticker','Date']).to_csv('stock_data.csv')
    
```

```python

import requests
import pandas as pd
from bs4 import BeautifulSoup

# Get the webpage
url = 'https://www.sec.gov/cgi-bin/browse-edgar?action=getcompany&CIK=0001045810&type=10-K&dateb=&owner=exclude&count=40'
response = requests.get(url)

# Load the webpage content
soup = BeautifulSoup(response.content, 'html.parser')
print(soup.prettify())

# Get the table having the filings data
table = soup.find('table', {'class': 'tableFile2'}) 

# Get all the rows of table
rows = table.find_all('tr') # find error here         

# Get the columns from the table
columns = [v.text.replace('\n', '') for v in rows[0].find_all('th')]
columns


# Initialize the dataframe with the columns
df = pd.DataFrame(columns=columns)


# Get all the table rows
for i in range(1, len(rows)):
    tds = rows[i].find_all('td')
    
    # Get all the table data
    data = [td.text.replace('\n', '').strip() for td in tds]
    
    # Append the data to the dataframe
    df = df.append(pd.Series(data=data, index=columns), ignore_index=True)

# Display the dataframe
print(df)

# Get the link to the XBRL instance document
instance_document = 'https://www.sec.gov/Archives/' + df['Filings'].iloc[0]

# Get the XBRL instance document
xbrl_response = requests.get(instance_document)

# Load the XBRL instance document
xbrl_response_content = BeautifulSoup(xbrl_response.content, 'lxml')

# Find and print stockholder's equity
stockholder_equity = xbrl_response_content.find('us-gaap:stockholdersequity')

print(stockholder_equity)

# Find and print the common stock value
common_stock = xbrl_response_content.find('us-gaap:commonstockvalue')

print(common_stock)

# Find and print the retained earnings value
retained_earnings = xbrl_response_content.find('us-gaap:retainedearningsdeficit')

print(retained_earnings)
```

```python
def format_premium_m(usd: float) -> str:
    """Format a USD amount to millions string (e.g., '$4.33M')."""
    if not usd or usd == 0:
        return "$0"
    if abs(usd) >= 1000_000_000_000:
        return f"${usd / 1000_000_000_000:.2f}T"
    if abs(usd) >= 1000_000_000:
        return f"${usd / 1000_000_000:.2f}B"
    if abs(usd) >= 1_000_000:
        return f"${usd / 1_000_000:.2f}M"
    elif abs(usd) >= 1_000:
        return f"${usd / 1_000:.1f}K"
    else:
        return f"${usd:,.0f}"
```

```python
def _safe_float(val, default=0.0) -> float:
    """Convert any value to float, returning default on None/NaN/inf."""
    try:
        v = float(val)
        return default if (math.isnan(v) or math.isinf(v)) else v
    except (TypeError, ValueError):
        return default


def _safe_interest_coverage(ebit: float, interest_expense) -> float:
    if ebit <= 0:
        return -1.0   # operating at a loss sentinel
    ie = _safe_float(interest_expense)
    if abs(ie) < 1_000:
        return 999.0  # debt-free sentinel
    return ebit / abs(ie)


def _safe_peg(pe: float, peg_raw, cagr_3yr: float) -> float:
    raw = _safe_float(peg_raw)
    if raw > 0:
        return raw
    if pe > 0 and cagr_3yr > 0:
        return pe / (cagr_3yr * 100)
    return None


def _apply_scoring(yoy, cagr, margin_trend, fcf_ttm, net_cash, pe) -> int:
    score = 0
    if yoy > 0.15:                     score += 1
    if cagr > 0.10:                    score += 1
    if margin_trend.startswith("EXP"): score += 1
    if fcf_ttm > 0:                    score += 1
    if net_cash > 0:                   score += 1
    if pe and pe > 0 and pe > (1 / RISK_FREE_RATE):
        score = max(0, score - 1)
    return score


def _verdict(score: int, flags: list) -> str:
    flags = flags or []  # guard against None
    warnings  = [f for f in flags if f.startswith("⚠️") or f.startswith("🔴")]
    red_flags = [f for f in flags if f.startswith("🔴")]
    base = "STRONG BUY" if score >= 4 else "BUY" if score >= 2 else "HOLD"
    if base == "STRONG BUY" and (len(warnings) >= 3 or len(red_flags) >= 1):
        return "BUY WITH CAUTION ⚠️"
    return base
```

```python
def get_yfinance_fallback(ticker: str) -> dict:
    try:
        print(f"--- Starting yfinance fallback for {ticker} ---")
        stock = yf.Ticker(ticker)
        print(f"Ticker object initialized")
        
        data = stock.get_info()

        # ── Use modern API: income_stmt rows are metrics, cols are dates ──
        inc_ttm  = stock.get_income_stmt(freq='trailing').T
        print(f"inc_ttm shape: {inc_ttm.shape}")

        inc_ann  = stock.get_income_stmt(freq='yearly').T
        print(f"inc_ann shape (pre-drop): {inc_ann.shape}")

        inc_ann = inc_ann.dropna(how='all')
        print(f"inc_ann shape (post-drop): {inc_ann.shape}")

        bal      = stock.get_balance_sheet(freq='yearly').T
        print(f"bal shape: {bal.shape}")

        cf_ttm   = stock.get_cashflow(freq='trailing').T
        print(f"cf_ttm shape: {cf_ttm.shape}")

        fi       = stock.fast_info
        print(f"fast_info accessed")

        if inc_ann.empty or len(inc_ann) < 4:
            print(f"FAILED: Insufficient history for {ticker}")
            return {"error": f"Insufficient historical data for {ticker}"}

        # ── Helper: first non-null value across candidate column names ────
        def col(df, *names, idx=0):
            for n in names:
                if n in df.columns:
                    v = df[n].iloc[idx]
                    f = _safe_float(v)
                    if f != 0.0:
                        return f
            return 0.0

        # ── Revenue & Growth (annual for CAGR, TTM for current) ──────────
        rev_ttm      = col(inc_ttm, 'TotalRevenue', 'Revenue')
        print(f"rev_ttm: {format_premium_m(rev_ttm)}")

        if rev_ttm == 0.0:
            rev_ttm  = col(inc_ann, 'TotalRevenue', 'Revenue', idx=0)
            print(f"rev_ttm (fallback to annual): {format_premium_m(rev_ttm)}")

        past_rev  = col(inc_ann, 'TotalRevenue', 'Revenue', idx=1)
        print(f"past_rev (1yr ago): {format_premium_m(past_rev)}")

        three_yr_rev = col(inc_ann, 'TotalRevenue', 'Revenue', idx=3)
        print(f"three_yr_rev (3yr ago): {format_premium_m(three_yr_rev)}")

        yoy_growth = (rev_ttm - past_rev) / past_rev if past_rev > 0 else 0.0
        print(f"yoy_growth: {yoy_growth:.2%}")

        cagr_3yr   = min(((rev_ttm / three_yr_rev) ** (1/3)) - 1, 9.99) if three_yr_rev > 0 else 0.0
        print(f"cagr_3yr: {cagr_3yr:.2%}")

        # ── Operating income — TTM preferred ─────────────────────────────
        op_inc_ttm  = col(inc_ttm, 'OperatingIncome', 'EBIT')
        print(f"op_inc_ttm: {format_premium_m(op_inc_ttm)}")

        op_inc_prev = col(inc_ann, 'OperatingIncome', 'EBIT', idx=1)
        print(f"op_inc_prev: {format_premium_m(op_inc_prev)}")

        rev_prev    = col(inc_ann, 'TotalRevenue', 'Revenue', idx=1)
        print(f"rev_prev: {format_premium_m(rev_prev)}")

        op_margin      = op_inc_ttm / rev_ttm  if rev_ttm  > 0 else 0.0
        print(f"op_margin: {op_margin:.2%}")

        prev_op_margin = op_inc_prev / rev_prev if rev_prev > 0 else 0.0
        print(f"prev_op_margin: {prev_op_margin:.2%}")

        margin_trend   = "EXPANDING 📈" if op_margin > prev_op_margin else "CONTRACTING 📉"
        print(f"margin_trend: {margin_trend}")

        # ── Net income ────────────────────────────────────────────────────
        net_inc = col(inc_ttm, 'NetIncome', 'NetIncomeCommonStockholders',
                      'NetIncomeIncludingNoncontrollingInterests')
        print(f"net_inc (initial): {format_premium_m(net_inc)}")

        if net_inc == 0.0:
            net_inc = col(inc_ann, 'NetIncome', 'NetIncomeCommonStockholders', idx=0)
            print(f"net_inc (fallback): {net_inc}")

        net_margin = net_inc / rev_ttm if rev_ttm > 0 else 0.0
        print(f"net_margin: {net_margin:.2%}")

        # ── EBIT for interest coverage ────────────────────────────────────
        ebit = col(inc_ttm, 'EBIT', 'OperatingIncome')
        print(f"ebit: {format_premium_m(ebit)}")

        # ── FCF — TTM ─────────────────────────────────────────────────────
        fcf = col(cf_ttm, 'FreeCashFlow', 'CapexAndFreeCashFlow')
        print(f"fcf: {format_premium_m(fcf)}")

        fcf_margin = fcf / rev_ttm if rev_ttm > 0 else 0.0
        print(f"fcf_margin: {fcf_margin:.2%}")

        fcf_conv   = fcf / net_inc  if net_inc  != 0 else 0.0
        print(f"fcf_conv: {fcf_conv:.2%}")

        # ── Balance sheet (most recent annual) ───────────────────────────
        total_debt   = col(bal, 'TotalDebt', 'LongTermDebt',
                           'LongTermDebtAndCapitalLeaseObligation')
        print(f"total_debt: {format_premium_m(total_debt)}")

        total_equity = col(bal, 'StockholdersEquity', 'CommonStockEquity',
                           'TotalEquityGrossMinorityInterest')
        print(f"total_equity: {format_premium_m(total_equity)}")

        cash         = col(bal, 'CashAndCashEquivalents',
                           'CashCashEquivalentsAndShortTermInvestments')
        print(f"cash: {format_premium_m(cash)}")

        curr_assets  = col(bal, 'CurrentAssets')
        print(f"curr_assets: {format_premium_m(curr_assets)}")

        curr_liab    = col(bal, 'CurrentLiabilities')
        print(f"curr_liab: {format_premium_m(curr_liab)}")

        net_cash      = cash - total_debt
        print(f"net_cash: {format_premium_m(net_cash)}")

        current_ratio = curr_assets / curr_liab if curr_liab > 0 else 0.0
        print(f"current_ratio: {current_ratio:.2f}")

        de_ratio      = total_debt / total_equity if total_equity != 0 else 0.0
        print(f"de_ratio: {de_ratio:.2f}")

        # ── ROIC & ROE ────────────────────────────────────────────────────
        nopat            = ebit * 0.79
        print(f"nopat: {format_premium_m(nopat)}")

        invested_capital = total_debt + total_equity
        print(f"invested_capital: {format_premium_m(invested_capital)}")

        roic = nopat / invested_capital if invested_capital != 0 else 0.0
        print(f"roic: {roic:.2%}")

        roe  = net_inc / total_equity   if total_equity != 0 else 0.0
        print(f"roe: {roe:.2%}")

        # ── Interest coverage ─────────────────────────────────────────────
        ie = col(inc_ttm, 'InterestExpense', 'InterestExpenseNonOperating',
                 'NetInterestIncome')
        print(f"interest_expense: {ie}")

        interest_coverage = _safe_interest_coverage(ebit, ie)
        print(f"interest_coverage: {interest_coverage}")

        # ── P/E — fast_info.market_cap is the most reliable source ───────
        mktcap = _safe_float(getattr(fi, 'market_cap', None))
        print(f"mktcap: {format_premium_m(mktcap)}")

        pe = mktcap / net_inc if (mktcap > 0 and net_inc != 0) else None
        print(f"pe: {pe}")

        # PEG: try info{} first, fall back to manual
        try:
            peg_raw = _safe_float(stock.info.get('pegRatio'))
            print(f"peg_raw from info: {peg_raw}")
        except Exception as e:
            print(f"peg_raw fetch failed: {e}")
            peg_raw = 0.0

        peg = _safe_peg(pe, peg_raw if peg_raw > 0 else None, cagr_3yr)
        print(f"final peg: {peg}")

        # ── Insider transactions (yfinance native) ────────────────────────
        buys = sells = 0
        try:
            txns = stock.insider_transactions
            if txns is not None and not txns.empty:
                buys  = int((txns['Transaction'].str.contains('Buy',  na=False)).sum())
                sells = int((txns['Transaction'].str.contains('Sell', na=False)).sum())
                print(f"insiders - buys: {buys}, sells: {sells}")
            else:
                print("insiders - no data")
        except Exception as e:
            print(f"insider check skipped: {e}")
            pass

        score = _apply_scoring(yoy_growth, cagr_3yr, margin_trend, fcf, net_cash, pe)
        print(f"Final Score calculated: {score}")

        return {
            "ticker":        ticker.upper(),
            "growth":        {"rev": rev_ttm, "yoy": yoy_growth, "cagr": cagr_3yr},
            "margins":       {"om": op_margin, "trend": margin_trend, "net": net_margin},
            "fcf":           {"val": fcf, "margin": fcf_margin, "yield": 0.0, "conversion": fcf_conv},
            "solvency":      {"de": de_ratio, "net_cash": net_cash, "cr": current_ratio,
                              "interest_coverage": interest_coverage},
            "profitability": {"roe": roe, "roic": roic},
            "valuation":     {"pe": pe, "peg": peg},
            "insiders":      {"buys": buys, "sells": sells},
            "score":         score,
            "source":        "yfinance (Fallback) ⚠️"
        }
    except Exception as e:
        print(f"CRITICAL ERROR in get_yfinance_fallback for {ticker}: {str(e)}")
        return {"error": f"Total Data Failure for {ticker}: {str(e)}"}
```

```python
def fetch_fmp_data(endpoint: str, ticker: str, params: Dict = {}) -> List[Dict]:
    """Helper to fetch from FMP using path-based symbols for the Stable API."""
    params["apikey"] = FMP_API_KEY
    symbol = dict(ticker)['symbol'].upper()
    # Construct URL as /stable/endpoint/TICKER
    url = f"{BASE_URL}/{endpoint}?symbol={symbol}"
    try:
        response = requests.get(url, params=params, timeout=10)
        if response.status_code != 200 or not response.json():
            return []
        return response.json()
    except Exception as e:
        print(f"DEBUG: FMP Request failed: {e}")
        return []
```

```python
def get_comprehensive_fundamentals(ticker: str) -> dict:
    ticker = ticker.upper()
    try:

        inc = fetch_fmp_data("income-statement", {"symbol": ticker, "limit": 5})
        print('Fetched from FMP: ', inc[1])
        if not inc or len(inc) < 4:
            return get_yfinance_fallback(ticker)

        bal      = fetch_fmp_data("balance-sheet-statement", {"symbol": ticker, "limit": 5})
        cf       = fetch_fmp_data("cash-flow-statement",     {"symbol": ticker, "limit": 5})
        ratios   = fetch_fmp_data("ratios-ttm",              {"symbol": ticker})
        metrics  = fetch_fmp_data("key-metrics-ttm",         {"symbol": ticker})
        insiders = fetch_fmp_data("insider-trading",         {"symbol": ticker, "limit": 25})

        if not bal or not cf or not ratios or not metrics:
            return get_yfinance_fallback(ticker)
    except Exception:
        return get_yfinance_fallback(ticker)

    curr, past, three_yr = inc[0], inc[1], inc[3]
    rev_ttm    = _safe_float(curr.get("revenue"))
    yoy_growth = (rev_ttm - _safe_float(past.get("revenue"))) / _safe_float(past.get("revenue"), 1)
    cagr_3yr   = min((( rev_ttm / _safe_float(three_yr.get("revenue"), 1)) ** (1/3)) - 1, 9.99)

    op_margin    = _safe_float(curr.get("operatingIncomeRatio"))
    margin_trend = "EXPANDING 📈" if op_margin > _safe_float(past.get("operatingIncomeRatio")) else "CONTRACTING 📉"

    fcf_ttm    = _safe_float(cf[0].get("freeCashFlow"))
    fcf_margin = fcf_ttm / rev_ttm if rev_ttm > 0 else 0.0
    fcf_yield  = _safe_float(metrics[0].get("freeCashFlowYieldTTM")) if metrics else 0.0
    net_income = _safe_float(curr.get("netIncome"), 1)
    fcf_conv   = fcf_ttm / net_income if net_income != 0 else 0.0

    total_debt   = _safe_float(bal[0].get("totalDebt"))
    total_equity = _safe_float(bal[0].get("totalStockholdersEquity"), 1)
    net_cash     = _safe_float(bal[0].get("cashAndCashEquivalents")) - total_debt
    current_ratio = _safe_float(ratios[0].get("currentRatioTTM")) if ratios else 0.0

    ebit             = _safe_float(curr.get("operatingIncome"))
    interest_expense = _safe_float(curr.get("interestExpense"))
    interest_coverage = _safe_interest_coverage(ebit, interest_expense)

    roic = _safe_float(metrics[0].get("roicTTM")) if metrics else 0.0
    roe  = _safe_float(ratios[0].get("returnOnEquityTTM")) if ratios else 0.0
    pe   = _safe_float(ratios[0].get("priceEarningsRatioTTM")) if ratios else 0.0
    peg  = _safe_peg(pe, ratios[0].get("priceEarningsToGrowthRatioTTM") if ratios else None, cagr_3yr)

    buys  = len([x for x in insiders if "Buy"  in x.get("typeOfTransaction", "")]) if insiders else 0
    sells = len([x for x in insiders if "Sell" in x.get("typeOfTransaction", "")]) if insiders else 0
    score = _apply_scoring(yoy_growth, cagr_3yr, margin_trend, fcf_ttm, net_cash, pe)

    return {
        "ticker":        ticker,
        "growth":        {"rev": rev_ttm, "yoy": yoy_growth, "cagr": cagr_3yr},
        "margins":       {"om": op_margin, "trend": margin_trend, "net": _safe_float(curr.get("netIncomeRatio"))},
        "fcf":           {"val": fcf_ttm, "margin": fcf_margin, "yield": fcf_yield, "conversion": fcf_conv},
        "solvency":      {"de": total_debt / total_equity, "net_cash": net_cash,
                          "cr": current_ratio, "interest_coverage": interest_coverage},
        "profitability": {"roe": roe, "roic": roic},
        "valuation":     {"pe": pe, "peg": peg},
        "insiders":      {"buys": buys, "sells": sells},
        "score":         score,
        "source":        "Financial Modeling Prep (Primary)"
    }
```
